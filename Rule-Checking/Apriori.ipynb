```python
import numpy as np
import pandas as pd
from efficient_apriori import apriori

data = pd.read_csv("Stabilise.csv", header = 0)

data.shape
```

```python
records = []
for i in range(0,77):
    records.append([str(data.values[i,j]) for j in range (0,11)])
```

```python
association_rules = apriori(records, min_support = 1, min_confidence = 1)
association_results = list(association_rules)
```
